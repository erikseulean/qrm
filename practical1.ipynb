```r
data = get.hist.quote("^DJI", quote="Close", start="1992-01-01", end="2022-01-25", retclass="zoo")
```

```r
plot(data)
years = substr(index(data), start=0, stop=4)
years1 = years[-1]
ticks = which( years != years1)
axis(1, at = years[ticks], label = years[ticks], tcl = -0.3)
```
